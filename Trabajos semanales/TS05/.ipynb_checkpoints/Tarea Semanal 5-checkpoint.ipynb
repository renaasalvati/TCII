```python
# Inicialización e importación de módulos

# Módulos externos
import numpy as np
import scipy.signal as sig
import matplotlib as mpl
import math

from matplotlib import pyplot as plt
from pytc2.sistemas_lineales import analyze_sys, pretty_print_lti, bodePlot, pzmap, tf2sos_analog, pretty_print_SOS

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 13

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

```python
#Datos de enunciado para filtro pasa altos
fc = 300
fz = 100

n = 3 #Orden calculado
a_max = 3 #Atenuación elegida 

#Normalización
nf = 2*np.pi*fc #norma de frecuencia

wc_n = (2*np.pi*fc)/nf
wz_n = (2*np.pi*fz)/nf

#Conversión de parámetros para un filtro pasa bajos equivalente
Wc_n = 1/wc_n
Wz_n = round(1/wz_n)
```

```python
#Transferencia normalizada del filtro pasabajos prototipo
z, p, k = sig.buttap(n)
num_lp_n, den_lp_n = sig.zpk2tf(z, p, k)

sos_lp_n = tf2sos_analog(num_lp_n, den_lp_n)

print("Transferencia normalizada del filtro pasa-bajos prototipo")
pretty_print_SOS(sos_lp_n)#, mode='omegayq') #factorizando

#Cero de transmisión de 0db
num_z = [1., 0, Wz_n**2]
den_z = [Wz_n**2]

print("Cero de transmisión de 0db en 100Hz")
pretty_print_lti(num_z, den_z)

num_lp_z_n = np.polymul(num_lp_n, num_z)
den_lp_z_n = np.polymul(den_lp_n, den_z)

sos_lp_z_n = tf2sos_analog(num_lp_z_n, den_lp_z_n)

print("Transferencia normalizada del filtro pasa-bajos prototipo con el cero de transmisión")
pretty_print_lti(num_lp_z_n, den_lp_z_n)
pretty_print_SOS(sos_lp_z_n, mode='omegayq') #factorizando

#Aplicación del núcleo de transformación
num_hp_n, den_hp_n = sig.lp2hp(num_lp_z_n, den_lp_z_n)

print("Transferencia normalizada del filtro pasa-altos con el cero de transmisión")
pretty_print_lti(num_hp_n, den_hp_n)
```

```python
pzmap(sig.TransferFunction(num_hp_n, den_hp_n))
```

```python
#Cálculo de valores normalizados de componentes
#Etapa segundo orden
R_ = 1

#R1²-9R1+9=0
x1 = (-9+np.sqrt(9**2-(4*1*-9)))/(2*1)
x2 = (-9-np.sqrt(9**2-(4*1*-9)))/(2*1)

if(x1 > 0):
    R1_ = round(x1, 3)
else:
    R1_ = round(x2, 3)

C1_ = round(1/R1_, 3)
R2_ = round(R1_**2, 3)

z = (1/(C1_**2*R2_))*(1/R2_-1/R1_)

print(f"{z}  ")

print(f"Valores de los componentes normalizados de la etapa de segundo orden: \nR = {R_}, R1 = {R1_}, R2_ = {R2_}, C1_ = {C1_}")

#Etapa primer orden

C2_ = 1/R_

print(f"\nValores de los componentes normalizados de la etapa de primer orden: \nR = {R_}, C2_ = {C2_}")
```

```python
#Desnormalización de los componentes
ni = 1*(10**3)
nf = 2*np.pi*fc

R = R_*ni
R1 = R1_*ni
R2 = R2_*ni

C1 = round(C1_/(ni*nf), 10)
C2 = round(C2_/(ni*nf), 10)

print(f"Valores de los componentes desnormalizados: \nR = {R}, R1 = {R1}, R2 = {R2}, C1 = {C1}, C2 = {C2}")
```
