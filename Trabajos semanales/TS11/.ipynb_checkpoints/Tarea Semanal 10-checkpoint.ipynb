```python
import sympy as sp

from pytc2.sintesis_dipolo import cauer_RC, foster, foster_zRC2yRC
from pytc2.dibujar import display, dibujar_puerto_entrada, dibujar_puerto_salida, dibujar_funcion_exc_abajo, dibujar_elemento_serie, dibujar_elemento_derivacion, dibujar_tanque_RC_derivacion, dibujar_espacio_derivacion, Capacitor, Inductor, Resistor, ResistorIEC
from pytc2.remociones import remover_polo_dc, remover_polo_jw, remover_valor_en_dc, remover_valor_en_infinito, remover_polo_sigma
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
from IPython.display import display, Markdown

s = sp.symbols('s', complex=True)
```

```python
Z = (s**2 + 6*s + 8)/(s**2 + 4*s + 3)

print_latex(a_equal_b_latex_s('Z(s)', Z))

sigma_RC1 = -6

Za, Z1 = remover_valor_en_dc(Z, sigma_zero = sigma_RC1)

print_subtitle('Primer remoción (parcial): resistencia en serie')
print_latex(a_equal_b_latex_s('Z_1', Z1))
print_subtitle('Impedancia resultante')
print_latex(a_equal_b_latex_s('Z_a', Za))
```

```python
Ya = 1/Za

Yb, YRC1, R1, C1 = remover_polo_sigma(Ya, sigma = sigma_RC1, isImpedance = False, isRC = True )

print_subtitle('Segunda remoción: Tanque RC en derivación')
print_latex(a_equal_b_latex_s('Y_{RC_1}', YRC1))
print_subtitle('Admitancia resultante')
print_latex(a_equal_b_latex_s('Y_b', Yb))
```

```python
Zb = 1/Yb

sigma_RC2 = sp.nsimplify(-7/2)

Zc, Z2 = remover_valor_en_dc(Zb, sigma_zero = sigma_RC2)

print_subtitle('Primer remoción (parcial): resistencia en serie')
print_latex(a_equal_b_latex_s('Z_2', Z2))
print_subtitle('Impedancia resultante')
print_latex(a_equal_b_latex_s('Z_c', Zc))
```

```python
Yc = 1/Zc

Yd, YRC2, R2, C2 = remover_polo_sigma(Yc, sigma = sigma_RC2, isImpedance = False, isRC = True )

print_subtitle('Tercer remoción: Tanque RC en derivación')
print_latex(a_equal_b_latex_s('Y_{RC_2}', YRC2))
print_subtitle('Admitancia resultante')
print_latex(a_equal_b_latex_s('Y_d', Yd))

Z3 = 1/Yd

print_subtitle('Cuarta remoción: resistencia en serie')
print_latex(a_equal_b_latex_s('Z_3', Z3))
```

```python
print_subtitle('Red resultante')

d = dibujar_puerto_entrada('', voltage_lbl=('+', '$V$', '-'), current_lbl= '$I$')
d, zz_lbl = dibujar_funcion_exc_abajo(d, 'Z(S)', Z, hacia_salida=True, k_gap_width=0.5)
d = dibujar_elemento_serie(d, Resistor, Z1)
d = dibujar_espacio_derivacion(d)
d = dibujar_tanque_RC_derivacion(d, sym_R_label=R1, capacitor_lbl=C1)
d = dibujar_espacio_derivacion(d)
d = dibujar_elemento_serie(d, Resistor, Z3)
d = dibujar_espacio_derivacion(d)
d = dibujar_tanque_RC_derivacion(d, sym_R_label=R2, capacitor_lbl=C2)
d = dibujar_espacio_derivacion(d)

d = dibujar_elemento_derivacion(d, Resistor, Z3)

display(d)
```

```python
import numpy as np
from pytc2.remociones import remover_polo_dc, remover_polo_infinito
```

```python
Z = (s**2 + s + 1)/((s**2 + 2*s + 5)*(s+1))

print_latex(a_equal_b_latex_s('Z(s)', Z))

Y = 1/Z

Ya, YC1 = remover_polo_infinito(Y)

print_subtitle('Primer remoción: capacitor C1 en derivación')
print_latex(a_equal_b_latex_s('C_1', YC1/s))
print_subtitle('Admitancia resultante')
print_latex(a_equal_b_latex_s('Y_a', Ya))
```

```python
Ginf = sp.limit(Ya, s, sp.oo)
G0 = sp.limit(Ya, s, 0)

# remuevo la menor admitancia
R1 = 1/np.min((Ginf, G0))
print_subtitle('Segunda remoción: resistencia R1 en derivación')
print_latex(a_equal_b_latex_s('R_1', R1))

Yb = sp.factor(sp.simplify(sp.expand(Ya - 1/R1)))
print_subtitle('Admitancia resultante')
print_latex(a_equal_b_latex_s('Y_b', Yb))
```

```python
Zb = 1/Yb

Zc, ZL1 = remover_polo_infinito(Zb)

print_subtitle('Tercer remoción: inductor L1 en serie')
print_latex(a_equal_b_latex_s('L_1', ZL1/s))
print_subtitle('Impedancia resultante')
print_latex(a_equal_b_latex_s('Z_c', Zc))
```

```python
Zinf = sp.limit(Zc, s, sp.oo)
Z0 = sp.limit(Zc, s, 0)

# remuevo la menor admitancia
R2 = np.min((Zinf, Z0))
print_subtitle('Cuarta remoción: resistencia R2 en serie')
print_latex(a_equal_b_latex_s('R_2', R2))

Zd = sp.factor(sp.simplify(sp.expand(Zc - R2)))
print_subtitle('Admitancia resultante')
print_latex(a_equal_b_latex_s('Z_d', Zd))
```

```python
Yd = 1/Zd

Ye, YC2 = remover_polo_infinito(Yd)

print_subtitle('Quinta remoción: capacitor C2 en derivación')
print_latex(a_equal_b_latex_s('C_2', YC2/s))
print_subtitle('Impedancia resultante')
print_latex(a_equal_b_latex_s('Y_e', Ye))
```

```python
R3 = 1/Ye
print_subtitle('Sexta remoción: resistencia R3 en derivación')
print_latex(a_equal_b_latex_s('R_3', R3))
```

```python
print_subtitle('Red resultante')

d = dibujar_puerto_entrada('', voltage_lbl = ('+', '$V$', '-'), current_lbl = '$I$')
d, zz_lbl = dibujar_funcion_exc_abajo(d, 'Z',  Z, hacia_salida = True, k_gap_width = 0.5)
d = dibujar_elemento_derivacion(d, Capacitor, YC1/s)
d = dibujar_espacio_derivacion(d)
d = dibujar_elemento_derivacion(d, Resistor, R1)
d = dibujar_elemento_serie(d, Inductor, ZL1/s)
d = dibujar_elemento_serie(d, Resistor, R2)
d = dibujar_elemento_derivacion(d, Capacitor, YC2/s)
d = dibujar_espacio_derivacion(d)
d = dibujar_elemento_derivacion(d, Resistor, R3)

display(d)
```
