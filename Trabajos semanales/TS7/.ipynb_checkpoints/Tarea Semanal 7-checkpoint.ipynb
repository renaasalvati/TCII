```python
import numpy as np
import scipy.signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt

from pytc2.sistemas_lineales import plot_plantilla, analyze_sys, bodePlot, pretty_print_lti

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 90 # dpi
fig_font_size = 13

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

```python
fc = 10**3
wc = 2*np.pi*fc

fs = 100*(10**3)
fs_ = fs/fc

num = [1]
den = [1, np.sqrt(2), 1]

numz, denz = sig.bilinear(num, den, fs_);

pretty_print_lti(num, den)
pretty_print_lti(numz, denz)

af = sig.TransferFunction(num, den)
df = sig.TransferFunction(numz, denz, dt = 1/fs_)

all_sys = []
analog_filters = []

all_sys.append(af)
analog_filters.append(af)

plt.figure(1)
plt.cla()

npoints = 100
w_nyq = 2*np.pi*fs/2

w, mag, _ = af.bode(npoints)
plt.plot(w/w_nyq, mag)

#analyze_sys([af, df], sys_name = ['Filtro analógico', 'Filtro digital']);
```

```python
fs = 10*(10**3)

numz, denz = sig.bilinear(num, den, fs);

df = sig.TransferFunction(numz, denz, dt = 1/fs)

analyze_sys([af, df], sys_name = ['Filtro analógico', 'Filtro digital']);
```

```python
fc = 6*10**3
wc = 2*np.pi*fc

fs = 100*(10**3)

num = [wc**2]
den = [1, np.sqrt(2)*wc, wc**2]

numz, denz = sig.bilinear(num, den, fs);

af = sig.TransferFunction(num, den)
df = sig.TransferFunction(numz, denz, dt = 1/fs)

analyze_sys([af, df], sys_name = ['Filtro analógico', 'Filtro digital']);
```
